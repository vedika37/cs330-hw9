# Blog Post

Avocados have jumped in popularity over recent decades, with sales rising dramatically (source). However, the price of avocados is something that fluctuates and varies during times of the year based on a number of affecting factors including season, region of sale, and size of the fruit.

To gain a better understanding of grocery prices as consumers, we wanted to try to predict the average price of avocados a week in advance. We feel that this problem is one that provides benefits to avocado consumers, since it provides valuable insights into price trends and allows them to make an informed decision to plan a grocery trip.

Data is an important part of any Machine Learning problem, and has a huge impact on the accuracy of the results. The data we used to base our predictions off of included information about the Hass avocado, the kind most popular in the USA, making up for 95% of the avocados eaten in the country (source). It included data from the Hass Avocado Board website from May 2018, with weekly information on retail data at a national level, such as, retail price, region, and number of units sold. To make this data more comprehensive and easier to interpret, we performed exploratory analysis and cleaning before delving further into the problem, for example, replacing the grocery store codes for kinds of Hass avocados with easy to understand descriptive names.

To get a deeper understanding of the data, we also performed some initial analysis:

First, we wanted to be able to tell how much the price actually fluctuates, (Figure 1 below).

The general trend shows that the price of Avocados across the US increased over time, with periods of sharp fluctuations, suggesting seasonal variability and differences in supply and demand.

Additionally, we can see that the spread of data is uniform in 7-day intervals as well, as noted in Figure 2, showing that we have a robust dataset.

(Figure 2)


Since the data was collected weekly and we want to predict prices for the following week, this problem is one of time series forecasting. Time series data naturally has a sequential structure and time-based ordering that is challenging to capture with Linear models since it exhibits non-linear and cyclical patterns that we need to preserve while performing our analysis. In order to help our model recognize these patterns, we use lag features, capturing the value of a variable from a previous step (here, the price of avocados from the previous week), and provide the model with a glimpse of previous trends. This is done because time series data has a ‘temporal dependence’ - observations close in time are ofteb correlated, and we can assume that the average avocado price this week will be similar to the previous weeks.

We chose the Ridge regression model, which works well when the data has many features and avoids overfitting by penalizing large coefficients. This makes it especially suitable for time series data, where we often have correlated features and want to ensure the model generalizes well to unseen data.

We split our data into a training and test set, maintaining the time based order and using data up to a certain date for training and the one after that for testing. This was done to see how our model will perform on unseen future data.

Next, we used preprocessing to ensure the data is clean and ready for the model to learn from. Here's what we did:
Numeric Features: Filled missing values (e.g., Total Volume, Small Bags) with the median and standardized them to make comparisons fair across features
Categorical Features: Converted categories like type and region into binary columns using one-hot encoding, so the model can work with them as numbers
Dropping Irrelevant Features: Removed columns like Date, which don't add value to predictions
Target Variable: Identified AveragePriceNextWeek as the value we want the model to predict, separating it from the input features

Finally, we set the hyperparameter alpha to 1. Hyperparameters control how a model learns, and in Ridge regression, alpha adjusts how much we penalize larger coefficients. Setting it to 1 helped us balance fitting the training data well while keeping the model simple enough to perform better on new, unseen data.

Taking a deeper look at the results, we decided to use a metric called R², which is the coefficient of determination, and allows us to calculate how well the variation in avocado prices is matched by our predictions. The R² score measures how well the model's predictions match the actual data, where 1 indicates a perfect fit and 0 means the model does not perform better than a simple average. For this problem, an R² of 0.85 on the train set and 0.80 on the test set suggests the model can predict prices on unseen data without overfitting significantly to the training data. The R² score provides a quantitative way to assess the model's performance. On a more qualitative note, it is a fairly interpretable way of assessing model performance as well, and easy to understand, the plot of actual and predicted values below (Figure 3) shows our results as well.


(Figure 3)


However, there are some important caveats to consider when we discuss our results.
Firstly, to predict the average price of avocados in the following week, we used the lag variable of price of avocados from the previous week. This would not take into account long-term trends. Another caveat is that it is resource intensive. To find the optimal hyperparameter values for a stronger model, we tuned the model by testing out multiple values of the hyperparameter, which is resource intensive, especially on a larger dataset. Additionally, our model doesn’t take into account several important factors such as weather or crop yield in a particular year, thus, it may be taking a too simplistic view of the overall prediction.

Overall, this project was one that was highly interesting to us, as it allowed us to apply concepts we had learned in lecture to a situation with real-world impact, and it was a great learning opportunity for us!
